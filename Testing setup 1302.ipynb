```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.integrate import odeint

%matplotlib inline
```

```python
"""Constants"""
eps0 = 8.854e-12 #[F/m]
eps_rel = 1 #relative permittivity
eps = eps0 * eps_rel
k = 1/(4*math.pi*eps)
e = 1.60217662e-19 #elementary charge [C]
h = 6.62607004e-34 #Planck's constant [Js]
m_e = 9.10938356e-31 #electron mass [kg]
```

```python
"""Initial parameter functions"""
def seperation_angle(n):
    return ( (2*math.pi)/n )

def setup_radius(sep, diam, PHI):
    return ( (sep + diam)/(2 * math.sin(PHI/2)) )

def cog_length( Rp, PHI, width):
    r = (Rp*math.sin(PHI) - width)/(2*math.sin(PHI/2))
    return r

def intertia( rho, thc, Rp):
    """Calculates inertia of inner rotating disk"""
    I = 0.5 * rho * math.pi * thc * ( Rp ** 4 ) #for disc, needs to be changed for cog shape
    return I

def arc_length( R, PHI):
    s = math.sin(PHI/2)
    return 2 * R * s
```

```python
"""Test lists"""
N = 1000
n_list = []
phi_list = []
R_list = []
arclength_list = []

"""Initial parameters:"""
'''QDs'''
diam = 10e-9 #diameter of QDs [m]
sep = 130e-9 #seperation of QDs [m] (corresponds to motor dead zone)
for n in np.arange(2, N):
    n_list.append(n)
    PHI = seperation_angle(n) #angle of seperation between QDs
    phi_list.append(PHI)
    #print(PHI)
    R = setup_radius(sep, diam, PHI)  #radius of QD setup [m] 5.36290830288e-07
    R_list.append(R)
    #print(R)
    arclength = arc_length( R, PHI)
    print(arclength)
    arclength_list.append(arclength)
```

```python
fig = plt.figure(figsize=(12, 9))
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)
ax1.set_xlabel('n', fontsize = '20')
ax1.set_ylabel('phi', fontsize = '20')
ax2.set_xlabel('n', fontsize = '20')
ax2.set_ylabel('R', fontsize = '20')
ax2.set_ylim( 0 , 3e-5)
ax3.set_xlabel('n', fontsize = '20')
ax3.set_ylabel('arc length', fontsize = '20')
ax3.set_ylim(0, 1e-9)

for i in np.arange(2, N):
    ax1.scatter(n_list[i], phi_list[i] , color = 'r')
    ax2.scatter(n_list[i], R_list[i] , color = 'g')
    ax3.scatter(n_list[i], arclength_list[i], color = 'b')
    #print (arclength_list[i])

plt.show()
```
